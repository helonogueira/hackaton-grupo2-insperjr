## Case AMBEV – Planejamento Long Neck Nordeste

### Objetivo
Avaliar os impactos produtivos, logísticos e financeiros do aumento de demanda de MALZBIER BRAHMA (+30% em fevereiro) e crescimento estrutural de +10% no total de Long Neck a partir de março.

A análise buscará responder:

- Devemos seguir com os incentivos comerciais?
- Qual será o plano de produção e transferência?
- Qual o impacto em custo?
- Quais os riscos envolvidos?

### Entendimento da Dor do Cliente

#### VP de Vendas
- Ganhar / manter market share
- Acelerar crescimento premium
- Evitar ruptura
- Proposta: +30% MALZBIER (Fev) e +10% LN (Mar+)

#### VP de Logística
- Capacidade produtiva no limite
- Restrição de envase
- Restrição de líquido (Goose)
- Lead time cabotagem: 25 dias
- Rodoviário: +60% custo e +5% avaria
- DOI mínimo: 12 dias

#### Dor da Companhia
Como crescer premium no Nordeste sem destruir margem ou aumentar risco operacional?
