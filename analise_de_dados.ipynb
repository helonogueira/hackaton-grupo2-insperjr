### 3. Análise de Nível de Serviço (DOI) – Cenário Base (W0)

Conforme premissa do case, o DOI mínimo exigido para a região Nordeste é de 12 dias. Na planilha é calculado a partir de:

DOI = Estoque Final / (Demanda da próxima semana / 6)

Relembrando o conceito de DOI: é o indicador que mostra por quantos dias o estoque atual consegue atender a demanda média, sem receber reposição.

Nesta etapa, analisamos o cenário base (antes do aumento de demanda) para identificar:

- Regiões já abaixo do DOI mínimo -> risco de ruptura; 
- Regiões em zona de risco (próximas de 12 dias);
- Possível fragilidade estrutural do sistema.

```python
df_w0 = df_w0.iloc[0:6]   # pega só as 6 primeiras regiões
df_w0[['GEO_REG', 'Suf_final_dias']]
```

- **DOI Total = 7,14 dias**, significativamente abaixo do mínimo estabelecido;
- Regiões críticas como:
  - Mapapi: 2,93 dias;
  - NO Centro: 1,16 dias;
  - NO Araguaia: 0 dias;
- Apenas NE Norte apresenta nível confortável (19,64 dias), indicando desequilíbrio regional na distribuição de estoque.

Observação: A região do Araguaia possui modelo logístico distinto, com retirada direta pelos revendedores na unidade de Uberlândia., dessa forma, o risco de ruptura nessa região não impacta diretamente a malha logística do Nordeste, devendo ser analisado separadamente.

Esse resultado evidencia que o produto que receberá um aumento de demanda de +30% já opera abaixo do nível mínimo de segurança.

Portanto, antes mesmo da aplicação do incentivo comercial, o sistema já apresenta risco elevado de ruptura, especialmente em determinadas regiões.

### 4. Análise de Aumento de Demanda (+30%) – MALZBIER

O aumento proposto no case prevê um crescimento de +30% na demanda do MALZBIER para fevereiro.

Considerando que o DOI é calculado como:

DOI = Estoque Final / (Demanda da próxima semana / 6)

Temos que o DOI é inversamente proporcional à demanda.  
Logo, mantendo o estoque constante, um aumento de 30% na demanda reduz o DOI aproximadamente na mesma proporção.

Aplicando essa variação ao cenário base:

- DOI Total atual: **7,14 dias**
- DOI estimado com +30%: aproximadamente **4,21 dias**

Regionalmente, a situação torna-se ainda mais crítica:

- Mapapi: cairia de 2,93 para cerca de 0,92 dias;
- NO Centro: cairia de 1,16 para menos de 1 dia;
- Regiões já abaixo do mínimo afastam-se ainda mais do nível de segurança.

Isso indica que, sem ajustes estruturais em produção ou transferências, o aumento de demanda ampliará significativamente o risco de ruptura no Nordeste.
