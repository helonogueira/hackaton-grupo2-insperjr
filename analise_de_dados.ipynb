### 4. Análise de Aumento de Demanda (+30%) – MALZBIER

O aumento proposto no case prevê um crescimento de +30% na demanda do MALZBIER para fevereiro.

Considerando que o DOI é calculado como:

DOI = Estoque Final / (Demanda da próxima semana / 6)

Temos que o DOI é inversamente proporcional à demanda.  
Logo, mantendo o estoque constante, um aumento de 30% na demanda reduz o DOI aproximadamente na mesma proporção.

Aplicando essa variação ao cenário base:

- DOI Total atual: **7,14 dias**
- DOI estimado com +30%: aproximadamente **4,21 dias**

Regionalmente, a situação torna-se ainda mais crítica:

- Mapapi: cairia de 2,93 para cerca de 0,92 dias;
- NO Centro: cairia de 1,16 para menos de 1 dia;
- Regiões já abaixo do mínimo afastam-se ainda mais do nível de segurança.

Isso indica que, sem ajustes estruturais em produção ou transferências, o aumento de demanda ampliará significativamente o risco de ruptura no Nordeste.

### 6. Teste de Sustentabilidade Operacional: Limite Máximo via Redistribuição

O DOI máximo possível sob redistribuição perfeita é dado por:

DOI_max = EF_total / (Demanda_total / 6)

No cenário com nova demanda, esse valor é aproximadamente 4,53 dias.

Isso significa que, mesmo sob alocação perfeitamente eficiente entre as regiões, o sistema permanece significativamente abaixo do mínimo exigido de 12 dias.

```python
# DOI maximo possível por distribuição
ef_total = df_novo.loc[df_novo["GEO_REG"]=="TOTAL","EF"].values[0]
demanda_total = df_novo.loc[df_novo["GEO_REG"]=="TOTAL","Demanda"].values[0]

doi_uniforme = ef_total / (demanda_total/6)

doi_uniforme
```

```python
df_novo["EF_uniforme"] = doi_uniforme * (df_novo["Demanda"]/6)
df_novo["Redistribuicao_Necessaria"] = df_novo["EF_uniforme"] - df_novo["EF"]

df_novo[[
    "GEO_REG",
    "EF",
    "EF_uniforme",
    "Redistribuicao_Necessaria"
]]
```

A redistribuição necessária indica que:

- NE Norte e NE Sul possuem estoque excedente relativo;
- Mapapi e NO Centro demandariam volumes adicionais relevantes para equalização;
- Ainda assim, o nível uniforme alcançado seria de apenas 4,5 dias.

Portanto, a redistribuição interna, isoladamente, não é suficiente para estabilizar o sistema.
